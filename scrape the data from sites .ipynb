```python
# import required library
import requests
from bs4 import BeautifulSoup
```

```python
# from this site
url = "https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States"
```

```python
res  = requests.get(url)
```

```python
res.status_code # to check request it ok or not
```

```python
soup = BeautifulSoup(res.text,'html')
```

```python
soup.title.string
```

```python
my_table = soup.find('table',attrs={"class":"wikitable sortable plainrowheaders"})
```

```python
my_table
```

```python
# select or find all "th" tag
th_tags = my_table.find_all('th')
names = []
for item in th_tags:
    a_links = item.find_all('a')
    for i in a_links:
        names.append(i.string)
print(names)
```

```python
final_list = names[9:]
```

```python
states = []
for str in final_list:
    if len(str) > 3:
        states.append(str)
        
print(states)
```

```python
divs = my_table.find_all("div")
```

```python
pop = []

for i in divs :
    pop.append(i.string)
pop
```

```python
# remove unwanted data
pop_final = []
for i in pop:
    if len(i) > 3:
        pop_final.append(i)
        
print(pop_final)
```

```python
div_repres = my_table.find_all("div")
```

```python
# remove unwanted data
repres =[]
for i in div_repres:
    #print(i)
    if len(i.string)<=3:
        
        repres.append(i.string)
```

```python
print(repres)
```

```python
admission = my_table.select('span')
```

```python
rectification = []
for i in admission:
    #print(i.string)
    if i.string != None:
        rectification.append(i.string)
        
print(rectification)
```

```python
len(pop_final) == len(states) ==len(rectification) ==len(repres)
```

```python
# writing data to csv
import pandas as pd

df = pd.DataFrame()
```

```python
df['state'] = states
df['population'] = pop_final
df['representative'] = repres
df['rectification'] = rectification
print(df)
df
```

```python
df.to_csv('us_info.csv') # to save data as csv file for further use
```
